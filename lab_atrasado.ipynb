```python
import numpy as np
import matplotlib.pyplot as plt

P0 = np.array([1, 1])
v = np.array([4, 3])

def encontrar_posicao(t):
    return P0 + t * v

t = 5
P_final = encontrar_posicao(t)

print("Posição final:", P_final)


plt.scatter(P0[0], P0[1], label="P0 (início)")
plt.scatter(P_final[0], P_final[1], label="P(t=5)", marker='x', s=100)
plt.plot([P0[0], P_final[0]], [P0[1], P_final[1]])
plt.legend()
plt.grid()
plt.show()
```

```python
import numpy as np

def calcular_distancia_ponto_reta(eq, ponto):
    A, B, C = eq
    x0, y0 = ponto
    return abs(A*x0 + B*y0 + C) / np.sqrt(A**2 + B**2)

dist = calcular_distancia_ponto_reta((3, 4, -12), (10, 10))
print("Distância ponto–reta:", dist)
```

```python
import sympy as sp

def resolver_intersecao(centro, r):
    x = sp.symbols('x')
    y = x
    h, k = centro

    eq = (x - h)**2 + (y - k)**2 - r**2
    sol = sp.solve(eq, x)

    if len(sol) == 0:
        return "Sem interseção"
    elif len(sol) == 1:
        return "Tangencia"
    else:
        return f"2 pontos de interseção: {sol}"

print(resolver_intersecao((10,10), 3))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

P1 = np.array([3,0,0])
P2 = np.array([0,4,0])
P3 = np.array([0,0,5])
Q  = np.array([5,5,5])

def plano(P1, P2, P3):
    n = np.cross(P2-P1, P3-P1)
    A, B, C = n
    D = -np.dot(n, P1)
    return A, B, C, D

def dist_ponto_plano(A,B,C,D, Q):
    x,y,z = Q
    return abs(A*x+B*y+C*z+D)/np.sqrt(A**2+B**2+C**2)

A,B,C,D = plano(P1,P2,P3)
dist = dist_ponto_plano(A,B,C,D, Q)

print("Equação do plano:", A, "x +", B, "y +", C, "z +", D, "= 0")
print("Distância Q–plano:", dist)


fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

tri_x = [P1[0],P2[0],P3[0],P1[0]]
tri_y = [P1[1],P2[1],P3[1],P1[1]]
tri_z = [P1[2],P2[2],P3[2],P1[2]]

ax.plot(tri_x, tri_y, tri_z)
ax.scatter(Q[0],Q[1],Q[2], color='red')
plt.show()
```
